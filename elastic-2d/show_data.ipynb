# Show Snapshot Data 

```python
import numpy as np
import matplotlib.pyplot as plt

def plot2d(img,nx,nz, title):
        dx, dz = 10*1e-3, 10*1e-3
        plt.figure(figsize=(5,5))
        plt.imshow(img[:,:].T,cmap='gray',aspect='auto',vmin=-0.05,vmax=0.05)
        plt.yticks(np.arange(0,nz,100), dx*np.arange(0,nx,100))
        plt.xticks(np.arange(0,nx,100), dz*np.arange(0,nz,100))
        plt.xlabel("Distance(km)")
        plt.ylabel("Depth(km)")
        plt.title("%s"%title)
        

nx , nz = 400, 400
savefig = False
filename0 = "/home/lcl1/code/elastic-2d-forward/output/junyun_vx.dat"
filename1 = "/home/lcl1/code/elastic-2d-forward/output/junyun_vz.dat"
# filename2 = "/home/lcl1/code/elastic-2d-forward/output/junyun_vxp.dat"
# filename3 = "/home/lcl1/code/elastic-2d-forward/output/junyun_vxs.dat"
figout_dir = "/home/lcl1/code/elastic-2d-forward/output/Fig/"

img0 = np.fromfile(filename0, 'f4').reshape(nx, nz)
img1 = np.fromfile(filename1, 'f4').reshape(nx, nz)
# img2 = np.fromfile(filename2, 'f4').reshape(nx, nz)
# img3 = np.fromfile(filename3, 'f4').reshape(nx, nz)
plot2d(img0, nx, nz, 'Snapshot-vx-pml surface')
if (savefig == True):
        plt.savefig(figout_dir+'snapshot-vx-pml surface.jpeg', dpi=300)
plot2d(img1, nx, nz, 'Snapshot-vz-pml surface')
if (savefig == True):
        plt.savefig(figout_dir+'snapshot-vz-pml surface.jpeg', dpi=300)

# plot2d(img2, nx, nz, 'Snapshot-vxp')
# if (savefig == True):
#         plt.savefig(figout_dir+'snapshot-vxp.jpeg', dpi=300)
# plot2d(img3, nx, nz, 'Snapshot-vxs')
# if (savefig == True):
#         plt.savefig(figout_dir+'snapshot-vxs.jpeg', dpi=300)

```

# Show obs data

```python
import matplotlib.pyplot as plt
import numpy as np

def plot2dobs(img,nx,nt, title):
        dx, dt = 10*1e-3, 1
        plt.figure(figsize=(5,5))
        plt.subplot(1,1,1)
        plt.imshow(img[:,:],cmap='gray',aspect='auto', vmin=-0.05, vmax=0.05)
        plt.yticks(np.arange(0,nt,100), dt*np.arange(0,nt,100))
        plt.xticks(np.arange(0,nx,100), dx*np.arange(0,nx,100))
        plt.xlabel("Distance(km)")
        plt.ylabel("Two Way Time(ms)")
        plt.title("%s"%title)

filename = "/home/lcl1/code/elastic-2d-forward/output/junyun_obsvx.dat"
filename1 = "/home/lcl1/code/elastic-2d-forward/output/junyun_obsvz.dat"
figout_dir = "/home/lcl1/code/elastic-2d-forward/output/Fig/"
nt, nx = 1000, 400
savefig = False

img = np.fromfile(filename, 'f4').reshape(nt, nx)
img1 = np.fromfile(filename1, 'f4').reshape(nt, nx)
plot2dobs(img[:,:], nx, nt, "seismogram_vx_pml")
if (savefig == True):
        plt.savefig(figout_dir+'seismogram-vx_pml.jpeg', dpi=300)

plot2dobs(img1[:,:], nx, nt, "seismogram_vz_pml")
if (savefig == True):
        plt.savefig(figout_dir+'seismogram-vz_pml.jpeg', dpi=300)
```
